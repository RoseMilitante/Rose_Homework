## Pymaceuticals Instructions

While your data companions rushed off to jobs in finance and government, you remained adamant that science was the way for you. Staying true to your mission, you've since joined Pymaceuticals Inc., a burgeoning pharmaceutical company based out of San Diego, CA. Pymaceuticals specializes in drug-based, anti-cancer pharmaceuticals. In their most recent efforts, they've since begun screening for potential treatments to squamous cell carcinoma (SCC), a commonly occurring form of skin cancer.

As their Chief Data Analyst, you've been given access to the complete data from their most recent animal study. In this study, 250 mice were treated through a variety of drug regimes over the course of 45 days. Their physiological responses were then monitored over the course of that time. Your objective is to analyze the data to show how four treatments (Capomulin, Infubinol, Ketapril, and Placebo) compare.

To do this you are tasked with:

* Creating a scatter plot that shows how the tumor volume changes over time for each treatment.
* Creating a scatter plot that shows how the number of [metastatic](https://en.wikipedia.org/wiki/Metastasis) (cancer spreading) sites changes over time for each treatment.
* Creating a scatter plot that shows the number of mice still alive through the course of treatment (Survival Rate)
* Creating a bar graph that compares the total % tumor volume change for each drug across the full 45 days.
* Include 3 observations about the results of the study. Use the visualizations you generated from the study data as the basis for your observations.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
data_complete = pd.merge(clinical_data, mouse_data, how="outer", on="Mouse ID")

# Display the data table for preview
data_complete.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
tumor_volume = data_complete.groupby(['Drug', 'Timepoint'])
tumor_df = tumor_volume["Tumor Volume (mm3)"].mean().to_frame()

# Preview DataFrame
tumor_df.head(20)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
tumor_stand_error = tumor_volume["Tumor Volume (mm3)"].sem().to_frame()

# Preview DataFrame

tumor_SEs = tumor_stand_error['Tumor Volume (mm3)'].unstack(0)
tumor_SEs
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_reformatted = tumor_df.unstack(0)

# Preview that Reformatting worked
tumor_reformatted
```

```python
#Here's my tumor means
capomulin = tumor_reformatted['Tumor Volume (mm3)'] ['Capomulin']
ceftamin = tumor_reformatted['Tumor Volume (mm3)'] ['Ceftamin']
infubinol = tumor_reformatted['Tumor Volume (mm3)'] ['Infubinol']
ketapril = tumor_reformatted['Tumor Volume (mm3)'] ['Ketapril']
naftisol = tumor_reformatted['Tumor Volume (mm3)'] ['Naftisol']
placebo = tumor_reformatted['Tumor Volume (mm3)'] ['Placebo']
propriva = tumor_reformatted['Tumor Volume (mm3)'] ['Propriva']
ramicane = tumor_reformatted['Tumor Volume (mm3)'] ['Ramicane']
stelasyn = tumor_reformatted['Tumor Volume (mm3)'] ['Stelasyn']
zoniferol = tumor_reformatted['Tumor Volume (mm3)'] ['Zoniferol']

#Here's my tumor standard errors
capo_se = tumor_SEs['Capomulin']
ceft_se = tumor_SEs['Ceftamin']
infu_se = tumor_SEs['Infubinol']
keta_se = tumor_SEs['Infubinol']
naft_se = tumor_SEs['Naftisol']
plac_se = tumor_SEs['Placebo']
prop_se = tumor_SEs['Propriva']
rami_se = tumor_SEs['Ramicane']
stel_se = tumor_SEs['Stelasyn']
zoni_se = tumor_SEs['Zoniferol']
```

```python
x_axis = [0,5,10,15,20,25,30,35,40,45]
x_limit = 60
plt.figure(figsize=(12,5))

plt.ylim(34, 73)
plt.xlim(-.5, 46)

plt.title("Tumor Response to Treatment", fontsize=15)
plt.xlabel("Time (Days)", fontsize=10)
plt.ylabel("Tumor Volume (mm3)", fontsize=10)
plt.grid()

# Generate the Plot (with Error Bars)
plt.errorbar(x_axis, capomulin, yerr=capo_se, fmt="o", color='red', ls='--', label='Capomulin')
#plt.errorbar(x_axis, ceftamin, yerr=ceft_se, fmt='x', color='orange', ls='--', label='Ceftamin')
plt.errorbar(x_axis, infubinol, yerr=infu_se, fmt='^', color='blue', ls='--', label='Infubinol')
plt.errorbar(x_axis, ketapril, yerr=keta_se, fmt='s', color='green', ls='--', label='Ketapril')
#plt.errorbar(x_axis, naftisol, yerr=naft_se, fmt='<', color='purple', ls='--', label='Naftisol')
plt.errorbar(x_axis, placebo, yerr=plac_se, fmt='d', color='black', ls='--', label='Placebo')
#plt.errorbar(x_axis, propriva, yerr=prop_se, fmt='h', color='magenta', ls='--', label='Propriva')
#plt.errorbar(x_axis, ramicane, yerr=rami_se, fmt='>', color='brown', ls='--', label='Ramicane')
#plt.errorbar(x_axis, stelasyn, yerr=stel_se, fmt='*', color='aqua', ls='--', label='Stelasyn')
#plt.errorbar(x_axis, zoniferol, yerr=zoni_se, fmt='p',color='grey', ls='--', label='Zoniferol')

plt.legend(bbox_to_anchor=(1, 0.5), loc='right')


plt.savefig("Tumor_Response.png")

# Save the Figure
plt.tight_layout
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 

meta = data_complete.groupby(['Drug', 'Timepoint'])
meta_df = meta["Metastatic Sites"].mean().to_frame()

# Preview DataFrame
meta_df.head(10)
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
meta_stand_error = meta["Metastatic Sites"].sem().to_frame()

# Preview DataFrame
meta_SEs = meta_stand_error['Metastatic Sites'].unstack(0)
meta_SEs
```

```python
# Minor Data Munging to Re-Format the Data Frames
meta_reformatted = meta_df.unstack(0)

# Preview that Reformatting worked
meta_reformatted
```

```python
#Here's my Metastatic means
meta_capomulin = meta_reformatted['Metastatic Sites'] ['Capomulin']
meta_ceftamin = meta_reformatted['Metastatic Sites'] ['Ceftamin']
meta_infubinol = meta_reformatted['Metastatic Sites'] ['Infubinol']
meta_ketapril = meta_reformatted['Metastatic Sites'] ['Ketapril']
meta_naftisol = meta_reformatted['Metastatic Sites'] ['Naftisol']
meta_placebo = meta_reformatted['Metastatic Sites'] ['Placebo']
meta_propriva = meta_reformatted['Metastatic Sites'] ['Propriva']
meta_ramicane = meta_reformatted['Metastatic Sites'] ['Ramicane']
meta_stelasyn = meta_reformatted['Metastatic Sites'] ['Stelasyn']
meta_zoniferol = meta_reformatted['Metastatic Sites'] ['Zoniferol']


#Here's my Metastatic standard errors
m_capo_se = meta_SEs['Capomulin']
m_ceft_se = meta_SEs['Ceftamin']
m_infu_se = meta_SEs['Infubinol']
m_keta_se = meta_SEs['Infubinol']
m_naft_se = meta_SEs['Naftisol']
m_plac_se = meta_SEs['Placebo']
m_prop_se = meta_SEs['Propriva']
m_rami_se = meta_SEs['Ramicane']
m_stel_se = meta_SEs['Stelasyn']
m_zoni_se = meta_SEs['Zoniferol']
```

```python
x_axis = [0,5,10,15,20,25,30,35,40,45]
x_limit = 60
plt.figure(figsize=(15,10))

plt.ylim(-0.5, 3.75)
plt.xlim(-1, 46)

plt.title("Metastatic Spread During Treatment", fontsize=15)
plt.xlabel("Treatment Duration (Days)", fontsize=10)
plt.ylabel("Metastatic Sites", fontsize=10)
plt.grid()

# Generate the Plot (with Error Bars)
plt.errorbar(x_axis, meta_capomulin, yerr=m_capo_se, fmt="o", color='red', ls='--', label='Capomulin')
#plt.errorbar(x_axis, meta_ceftamin, yerr=m_ceft_se, fmt='x', color='orange', ls='--', label='Ceftamin')
plt.errorbar(x_axis, meta_infubinol, yerr=m_infu_se, fmt='^', color='blue', ls='--', label='Infubinol')
plt.errorbar(x_axis, meta_ketapril, yerr=m_keta_se, fmt='s', color='green', ls='--', label='Ketapril')
#plt.errorbar(x_axis, meta_naftisol, yerr=m_naft_se, fmt='<', color='purple', ls='--', label='Naftisol')
plt.errorbar(x_axis, meta_placebo, yerr=m_plac_se, fmt='d', color='black', ls='--', label='Placebo')
#plt.errorbar(x_axis, meta_propriva, yerr=m_prop_se, fmt='h', color='magenta', ls='--', label='Propriva')
#plt.errorbar(x_axis, meta_ramicane, yerr=m_rami_se, fmt='>', color='brown', ls='--', label='Ramicane')
#plt.errorbar(x_axis, meta_stelasyn, yerr=m_stel_se, fmt='*', color='aqua', ls='--', label='Stelasyn')
#plt.errorbar(x_axis, meta_zoniferol, yerr=m_zoni_se, fmt='p',color='grey', ls='--', label='Zoniferol')

plt.legend(loc='best') 

plt.savefig("Metastatic_Spread.png")

# Save the Figure
plt.tight_layout
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
s = data_complete.groupby(['Drug', 'Timepoint'])
survival_df = s["Mouse ID"].count().to_frame()
survival_df = survival_df.rename(columns={"Mouse ID": "Mouse Count"})

# Preview DataFrame
survival_df.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_reformatted = survival_df.unstack(0)

# Preview that Reformatting worked
survival_reformatted
```

```python
mouse_capomulin = survival_reformatted['Mouse Count'] ['Capomulin']
mouse_ceftamin = survival_reformatted['Mouse Count'] ['Ceftamin']
mouse_infubinol = survival_reformatted['Mouse Count'] ['Infubinol']
mouse_ketapril = survival_reformatted['Mouse Count'] ['Ketapril']
mouse_naftisol = survival_reformatted['Mouse Count'] ['Naftisol']
mouse_placebo = survival_reformatted['Mouse Count'] ['Placebo']
mouse_propriva = survival_reformatted['Mouse Count'] ['Propriva']
mouse_ramicane = survival_reformatted['Mouse Count'] ['Ramicane']
mouse_stelasyn = survival_reformatted['Mouse Count'] ['Stelasyn']
mouse_zoniferol = survival_reformatted['Mouse Count'] ['Zoniferol']
```

```python
#turn survival rate into percentages
capomulin_rate = [(x/mouse_capomulin[0])*100 for x in mouse_capomulin]
ceftamin_rate = [(x/mouse_ceftamin[0])*100 for x in mouse_ceftamin]
infubinol_rate = [(x/mouse_infubinol[0])*100 for x in mouse_infubinol]
ketapril_rate = [(x/mouse_ketapril[0])*100 for x in mouse_ketapril]
naftisol_rate = [(x/mouse_naftisol[0])*100 for x in mouse_naftisol]
placebo_rate = [(x/mouse_placebo[0])*100 for x in mouse_placebo]
propriva_rate = [(x/mouse_propriva[0])*100 for x in mouse_propriva]
ramicane_rate = [(x/mouse_ramicane[0])*100 for x in mouse_ramicane]
stelasyn_rate = [(x/mouse_stelasyn[0])*100 for x in mouse_stelasyn]
zoniferol_rate = [(x/mouse_zoniferol[0])*100 for x in mouse_zoniferol]
```

```python
# Generate the Plot (Accounting for percentages)
x_axis = [0,5,10,15,20,25,30,35,40,45]
x_limit = 60
plt.figure(figsize=(15,10))

plt.ylim(30, 105)
plt.xlim(-1, 46)

plt.title("Survival During Treatment", fontsize=15)
plt.xlabel("Time (Days)", fontsize=10)
plt.ylabel("Survival Rate (%)", fontsize=10)
plt.grid()

# Generate the Plot (with Error Bars)
plt.errorbar(x_axis, capomulin_rate, fmt="o", color='red', ls='--', label='Capomulin')
#plt.errorbar(x_axis, ceftamin_rate, fmt='x', color='orange', ls='--', label='Ceftamin')
plt.errorbar(x_axis, infubinol_rate, fmt='^', color='blue', ls='--', label='Infubinol')
plt.errorbar(x_axis, ketapril_rate, fmt='s', color='green', ls='--', label='Ketapril')
#plt.errorbar(x_axis, naftisol_rate, fmt='<', color='purple', ls='--', label='Naftisol')
plt.errorbar(x_axis, placebo_rate, fmt='d', color='black', ls='--', label='Placebo')
#plt.errorbar(x_axis, propriva_rate, fmt='h', color='magenta', ls='--', label='Propriva')
#plt.errorbar(x_axis, ramicane_rate, fmt='>', color='brown', ls='--', label='Ramicane')
#plt.errorbar(x_axis, stelasyn_rate, fmt='*', color='aqua', ls='--', label='Stelasyn')
#plt.errorbar(x_axis, zoniferol_rate, fmt='p',color='grey', ls='--', label='Zoniferol')

plt.legend(loc='best') 
plt.savefig("Survival_Rate.png")

# Save the Figure
plt.tight_layout
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
change_percentage = (tumor_reformatted.loc[45, :] - tumor_reformatted.loc[0, :])/tumor_reformatted.loc[0, :] * 100

# Display the data to confirm
nice_looking_data = round(change_percentage, 1)
nice_looking_data = nice_looking_data.reset_index(drop=True)
changes_to_display = nice_looking_data.map('{}%'.format)
```

```python
# Store all Relevant Percent Changes into a Tuple
data_in_tuple = tuple(zip(change_percentage['Tumor Volume (mm3)'].index, nice_looking_data))
data_in_tuple
```

```python
# Splice the data between passing and failing drugs
# passing in green and failing in red
color_choice = []
for i in data_in_tuple:
    if i[1] > 0:
        color_choice.append('red')
    else:
        color_choice.append('green')
color_choice
```

```python
tick_locations = [value for value in data_in_tuple[0]]
tick_locations
```

```python
#figure out y-limits
maximum = 0
minimum = 0
for y in data_in_tuple:
    if y[1] > maximum:
        maximum = y[1]
        
for z in data_in_tuple:
    if z[1] < minimum:
        minimum = z[1]
```

```python
drugs = []
for n in range(0, len(data_in_tuple)):
    drugs.append(data_in_tuple[n][0])
drugs
```

```python
#amounts to display in the bar graph
changes_for_y = []
for n in range(0, len(changes_to_display)):
    changes_for_y.append(nice_looking_data[n])
changes_for_y
```

```python
# Orient widths. Add labels, tick marks, etc. 
x_axis = np.arange(len(data_in_tuple))

# Create the ticks for our bar chart's x axis
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, drugs, rotation="vertical", size=8)

plt.title("Tumor Change Over 45 Day Treatment", fontsize=16, fontweight="bold")
plt.xlabel("Drug")
plt.ylabel("% Tumor Volume change")
plt.xlim(-0.5, len(data_in_tuple)-0.5)
plt.ylim(minimum-5, maximum+0.7)
plt.grid()

plt.bar(x_axis, changes_for_y, width = .9, color= color_choice, align="center")

#bar labels
count = 0
for value in changes_for_y:
    if value < 0:
        y_coor = value + 4
    else:
        y_coor = value - 9
    plt.text(count, y_coor, str(round(value, 1)) + '%', ha = 'center', color = 'black')
    count += 1

# Save the Figure
plt.savefig("Summary_Bar_Graph.png")

# Show the Figure
plt.tight_layout()
plt.show()
```

# Observations
 - Tumor volume reduction is seen when Capomulin was administered over the course of the treatment as compared with Infubinol, Ketapril and the Placebo. 

 - Use of either the Placebo or Ketapril led to a greater increase of metastatic tumor spread during the treatment as compared with Capomulin and Infubinol. 

 - Overall, Capomulin is a favorable treatment as compared with the Placebo, Ketapril and Infubinol drugs leading to tumor volume reduction, less metastatic spread and ultimately a greater survival rate.
